```python
## Dividing the data between X and Y
```

```python
y=df.quality
x=df.drop('quality',axis=1)
```

```python
## Balancing the data
```

```python
from imblearn.over_sampling import RandomOverSampler
ros=RandomOverSampler(random_state=42)
x_ros,y_ros=ros.fit_resample(x,y)
print('The shape of data target data before oversampling: ',y.shape)
print('The shape of data target data before oversampling: ',y_ros.shape)
```

```python
x_ros.shape
```

```python
y_ros.value_counts()
```

```python
## Target variable is balanced now
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=56)
```

```python
## Classifications algorithms
```

```python
from sklearn.ensemble import RandomForestClassifier , ExtraTreesClassifier , GradientBoostingClassifier, AdaBoostClassifier, BaggingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix ,roc_curve, accuracy_score
from sklearn.model_selection import cross_val_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
RFC=RandomForestClassifier()
RFC.fit(x_train,y_train)
predRFC=RFC.predict(x_test)
print(accuracy_score(y_test,predRFC))
print(confusion_matrix(y_test,predRFC))
print(classification_report(y_test,predRFC))
```

```python
print(RFC.score(x_train,y_train))
print(RFC.score(x_test,y_test))
```

```python
RFCscore=cross_val_score(RFC,x,y)
print(np.mean(RFCscore))
print('Difference between accuracy and crossv validation score-',accuracy_score(y_test,predRFC)-np.mean(RFCscore))
```

```python
LR=LogisticRegression
LR.fit(x_ros,y_ros,sample_weight=7)
predLR=LR.predict(x_test)
print(accuracy_score(y_test,predLR))
print(confusion_matrix(y_test,predLR))
print(classification_report(y_test,predLR))
```
